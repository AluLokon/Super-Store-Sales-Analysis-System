## TRAIN TEST SPLIT

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(ip,op,test_size=0.2)
```

## STANDARD SCALER

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = x_train.select_dtypes(include=['number'])
x_test = x_test.select_dtypes(include=['number'])
x_train_s = sc.fit_transform(x_train)
x_test_s = sc.fit_transform(x_test)
```

## MACHINE LEARNING MODEL

#### 1) LOGISTIC REGRESSION

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(x_train_s,y_train)
```

##### Prediction

```python
ypred = lr.predict(x_test_s)
ypred
```

```python
from sklearn.metrics import accuracy_score,recall_score,f1_score,precision_score
acc1 = accuracy_score(ypred,y_test)
re1 = recall_score(ypred,y_test,pos_label=0,zero_division=0)
f11 = f1_score(ypred,y_test,pos_label=0,zero_division=0)
pre1 = precision_score(ypred,y_test,pos_label=0,zero_division=0)
```

```python
print("Accuracy : ",acc1)
print("Recall : ",re1)
print("F1 Score : ",f11)
print("Precision : ",pre1)
```

```python
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay
cm = confusion_matrix(ypred,y_test)
cm
```

```python
cmd = ConfusionMatrixDisplay(cm)
cmd.plot()
plt.show()
```

## KNN

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=5)     
knn.fit(x_train_s,y_train)
```

```python
y1=knn.predict(x_test_s)
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score
acc= accuracy_score(y1,y_test)
recall=recall_score(y1,y_test,pos_label=0,zero_division=0)
f1=f1_score(y1,y_test,pos_label=0,zero_division=0)
precision=precision_score(y1,y_test,pos_label=0,zero_division=0)
```

```python
print('Accuracy:',acc)
print('Recall:',recall)
print('F1 score:',f1)
print('Precision:',precision)
```
